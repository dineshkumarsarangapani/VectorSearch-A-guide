# Basics of Vector algebra

## Vectors

In mathematics, a vector is a mathematical object that represents both magnitude and direction. It is used to describe quantities that have both a size (magnitude) and a direction in space.

Geometrically, a vector can be visualized as an arrow in space, where the length of the arrow represents the magnitude of the vector, and the direction in which the arrow points represents the direction of the vector.

See: https://www.youtube.com/watch?v=fNk_zzaMoSs

```python
import matplotlib.pyplot as plt
import numpy as np

class Vector:
    def __init__(self, coordinates):
        self.coordinates = coordinates

    def __str__(self):
        return str(self.coordinates)

    def __add__(self, other):
        if len(self.coordinates) != len(other.coordinates):
            raise ValueError("Vector dimensions must match")
        new_coordinates = [x + y for x, y in zip(self.coordinates, other.coordinates)]
        return Vector(new_coordinates)

    def __sub__(self, other):
        if len(self.coordinates) != len(other.coordinates):
            raise ValueError("Vector dimensions must match")
        new_coordinates = [x - y for x, y in zip(self.coordinates, other.coordinates)]
        return Vector(new_coordinates)

    def __mul__(self, scalar):
        new_coordinates = [x * scalar for x in self.coordinates]
        return Vector(new_coordinates)

    def __len__(self):
        return len(self.coordinates)

    def dot(self, other):
        if len(self.coordinates) != len(other.coordinates):
            raise ValueError("Vector dimensions must match")
        return sum(x * y for x, y in zip(self.coordinates, other.coordinates))

    def magnitude(self):
        return (sum(x**2 for x in self.coordinates))**0.5


# Example usage
vector1 = Vector([1, 2, 3])
vector2 = Vector([4, 5, 6])

print("Vector 1:", vector1)
print("Vector 2:", vector2)

# Magnitude
magnitude = vector1.magnitude()
print("Magnitude of Vector 1:", magnitude)

dimension = len(vector1)
print("Dimension of Vector 1:", dimension)
```

```python

# Addition
addition_result = vector1 + vector2
print("Addition:", addition_result)

# Plotting vector addition
plt.quiver(0, 0, vector1.coordinates[0], vector1.coordinates[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(0, 0, vector2.coordinates[0], vector2.coordinates[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
plt.quiver(0, 0, addition_result.coordinates[0], addition_result.coordinates[1], angles='xy', scale_units='xy', scale=1, color='g', label='Addition')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Addition')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Subtraction
subtraction_result = vector1 - vector2
print("Subtraction:", subtraction_result)
# Plotting vector subtraction
plt.quiver(0, 0, vector1.coordinates[0], vector1.coordinates[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(0, 0, vector2.coordinates[0], vector2.coordinates[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')
plt.quiver(0, 0, subtraction_result.coordinates[0], subtraction_result.coordinates[1], angles='xy', scale_units='xy', scale=1, color='g', label='Subtraction')

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Vector Subtraction')
plt.legend()
plt.grid(True)
plt.show()
```

# Dot Product

The dot product, also known as the scalar product or inner product, is a mathematical operation between two vectors that results in a scalar value. It is commonly denoted using a dot (·) or by writing the vectors next to each other without any operator.

The essence of the dot product lies in measuring the similarity or alignment between two vectors. It provides a quantitative measure of how much two vectors point in the same direction.

When the dot product of two vectors is positive, it indicates that they have a similar direction. A value of zero means the vectors are orthogonal, or perpendicular, to each other. And if the dot product is negative, it suggests that the vectors are pointing in opposite directions.





```python

# Dot product
dot_product_result = vector1.dot(vector2)
print("Dot Product:", dot_product_result)
# Plotting dot product
fig, ax = plt.subplots()
plt.quiver(0, 0, vector1.coordinates[0], vector1.coordinates[1], angles='xy', scale_units='xy', scale=1, color='r', label='Vector 1')
plt.quiver(0, 0, vector2.coordinates[0], vector2.coordinates[1], angles='xy', scale_units='xy', scale=1, color='b', label='Vector 2')

max_val = max(np.concatenate((vector1.coordinates, vector2.coordinates)))
ax.text(0, -max_val, f"Dot Product: {dot_product_result}", ha='center', fontsize=12)

plt.xlim(-10, 10)
plt.ylim(-10, 10)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Dot Product')
plt.legend()
plt.grid(True)
plt.show()

```

# Vector Spaces

A vector space is a mathematical structure that consists of a set of vectors along with defined operations of addition and scalar multiplication. It provides a framework for studying and analyzing vectors and their properties.

In simpler terms, a vector space is like a playground where vectors can exist and interact. It sets the rules for how vectors behave and allows us to perform operations on them in a consistent way.

Imagine you have a special area where vectors can exist. This area has certain rules that all vectors must follow. These rules ensure that you can add vectors together and multiply them by numbers in a consistent way.

The rules include things like:

* When you add two vectors together, the result is always another vector within the same area.
* When you multiply a vector by a number, the result is still a vector within the same area.
* You can add vectors in any order, and it won't change the result.
* There's a special vector called the zero vector that doesn't change any vector when added to it.
* Every vector has a special partner that, when added together, gives the zero vector.
* Multiplying a vector sum by a number is the same as multiplying each vector by the number and then adding them.
* Multiplying a sum of numbers by a vector is the same as multiplying each number by the vector and then adding them.

These rules make sure that the operations you perform on vectors always give you meaningful results, and that the vectors behave nicely together.

So, a vector space is like a playground with rules that govern how vectors can be added and multiplied by numbers. It provides a framework for studying and working with vectors in a consistent and organized way.


```python
import numpy as np
import matplotlib.pyplot as plt

# Define the vectors
v1 = np.array([2, 1])
v2 = np.array([-1, 2])
v3 = np.array([-1, -2])

# Define the scalar values for linear combinations
s = np.linspace(-2, 2, 10)

# Create a figure and axes
fig, ax = plt.subplots()

# Plot the vectors
ax.quiver(0, 0, v1[0], v1[1], angles='xy', scale_units='xy', scale=1, color='b', label='v1')
ax.quiver(0, 0, v2[0], v2[1], angles='xy', scale_units='xy', scale=1, color='r', label='v2')
ax.quiver(0, 0, v3[0], v3[1], angles='xy', scale_units='xy', scale=1, color='g', label='v3')

# Plot the linear combinations
for scalar in s:
    v = scalar * v1 +   v2
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='y', alpha=0.5)
    v = scalar * v3 +   v1
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='m', alpha=0.5)
    v = scalar * v2 +   v3
    ax.quiver(0, 0, v[0], v[1], angles='xy', scale_units='xy', scale=1, color='brown', alpha=0.5)


# Set the limits for the plot
ax.set_xlim([-4, 4])
ax.set_ylim([-4, 4])

# Set the aspect ratio of the plot to be equal
ax.set_aspect('equal', adjustable='box')

# Set labels for the axes
ax.set_xlabel('X')
ax.set_ylabel('Y')

# Add a legend
ax.legend()

# Set the plot title
ax.set_title('Vector Space')

# Show the plot
plt.show()
```

# Closure under vector addition

Closure under vector addition is one of the fundamental rules of a vector space. It states that if you take any two vectors from a vector space and perform vector addition on them, the result will always be another vector that belongs to the same vector space.

In simpler terms, closure under vector addition means that when you add two vectors together, the resulting vector should still be within the same set or space of vectors. In other words, you cannot add two vectors and end up with a completely different type of object that is not considered a vector.

For example, let's consider a two-dimensional vector space. If we have two vectors u = (1, 2) and v = (3, 4), the sum of these vectors u + v = (4, 6) is another vector that belongs to the same two-dimensional vector space. This satisfies the closure under vector addition rule because the result is still a valid vector within the same space.

On the other hand, an anti-example would be if we attempted to add the vectors u = (1, 2) and v = (3, 4), but instead of performing vector addition, we accidentally multiplied the vectors together: u * v. The result of this operation is not a valid vector in the same two-dimensional vector space. It violates the closure under vector addition rule because the result is not a vector within the same space.

Visualizing this rule in the example Python program provided earlier, you can see how the vector addition operation satisfies the closure property by generating a new vector within the same space.

```python
import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

import numpy as np
import matplotlib.pyplot as plt

# Rule 1: Closure under vector addition
def closure_example():
    u = np.array([1, 2])
    v = np.array([3, 4])
    w = u + v

    vectors = np.vstack([u, v, w])
    max_val = np.max(np.abs(vectors))

    plt.quiver(0, 0, u[0], u[1], angles='xy', scale_units='xy', scale=1, color='b')
    plt.quiver(u[0], u[1], v[0], v[1], angles='xy', scale_units='xy', scale=1, color='r')
    plt.quiver(0, 0, w[0], w[1], angles='xy', scale_units='xy', scale=1, color='g')

    plt.xlim(-max_val - 1, max_val + 1)
    plt.ylim(-max_val - 1, max_val + 1)

    plt.gca().set_aspect('equal', adjustable='box')
    plt.show()

```
